```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
train=pd.read_csv("res/Auction_master_train.csv")
train=train.loc[train['Auction_key']!=10,]
```

```python
train # 데이터 값 로드
```

```python
train.corr()['Hammer_price'].sort_values() #Hammerprice 예측 값을 돌림 -> 상관관계가 있는것들을 나열됨 -> 높은 것과 낮은것이 나옴
# 마이너스는 반비례, 플러스는 정비례 -> 규칙성을 확인하기 위해 minimum_sales_price 를 이용하여 구함 -> 최소 시작가가 가장 연관이 있기 떄문에, 
#즉 최소 시작가가 높으면 경매가가 높다
```

```python
train['real']=train['Hammer_price']/train['Minimum_sales_price'] # 비율로 나타내보기
```

```python
train.corr()['real'].sort_values() #비율값을 상관관계로 나타낸 것 Auction_count 와 어느정도 상관성이 있다 판단
```

```python
plt.scatter(train['Auction_count'], train['real']) 
# 분산도를 표현 -> 옥션 카운트(유찰) 횟수가 12  -> 세로축이 점점 떨어짐 -> 최소 경매가 보다 첫번쨰는 1.8배 정도 -> 유찰이 거듭될수록 그대로
```

```python
cop=train.copy()
cop['x']=pd.cut(train['point.x'],100)
cop['y']=pd.cut(train['point.y'],100)
cop1=cop.groupby(['x','y']).size().reset_index(name='count')
cop2=cop.groupby(['x','y'])['real'].mean().reset_index(name='real')
cop=pd.merge(cop1,cop2,on=['x','y'],how='left')

cop=cop.loc[cop['count']>=9]
area = cop.pivot("y", "x", "real")
plt.subplots(1, 1, figsize=(7, 5))
plt.title("real by area")
sns.heatmap(area, annot=False, xticklabels=False,yticklabels=False,cmap="Blues")
plt.show()
```

```python
plt.scatter(train['addr_dong'], train['real']) # 무의미 한 값
```

```python
plt.scatter(train['Apartment_usage'], train['real']) # 아파트 건물의 대표 용도 
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
enc = []

x_train = train[['addr_dong', 'Apartment_usage', 'Auction_count']] # 최종적으로 뽑은 데이터 값
x_train['addr_dong'] = ordinal_encoder.fit_transform(x_train[['addr_dong']])
enc.append(ordinal_encoder.categories_)
x_train['Apartment_usage'] = ordinal_encoder.fit_transform(x_train[['Apartment_usage']]) #범주형 데이터를 카테고리컬로 변경-> ordinalenco
enc.append(ordinal_encoder.categories_)

y_train = train['real']

model = RandomForestRegressor()
model.fit(x_train, y_train)

# real 데이터를 예측한다 -> RandomForest 기법을 사용(카테고리컬 데이터 예측값을 사용)
```

```python
import pickle
filename = 'model.pickle'
pickle.dump(model, open(filename, 'wb'))
```

```python
import numpy as np

with open('test.npy', 'wb') as f:
    np.save(f, enc)
```

```python
def to_cat(x, arr_cat): #카테고리 함수
    try:
        ret = np.where(arr_cat[0]==x)[0][0]
    except:
        ret = 0
    return ret
```

```python
# 모델 로드
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
# 설정값 로드
with open('test.npy', 'rb') as f:
    lt = np.load(f, allow_pickle=True)
```

```python
# 테스트
```

```python
test=pd.read_csv("res/Auction_master_test.csv")
sub=pd.read_csv("res/Auction_submission.csv")
```

```python
x_test = test[['addr_dong', 'Apartment_usage', 'Auction_count']]
x_test['addr_dong'] = test['addr_dong'].apply(lambda x : to_cat(x, lt[0]))
x_test['Apartment_usage'] = test['Apartment_usage'].apply(lambda x : to_cat(x, lt[1]))
x_test.head()
```

```python
# 예측 수행
```

```python
x_test['real'] = loaded_model.predict(x_test)

x_test['Minimum_sales_price'] = test['Minimum_sales_price']
x_test['Hammer_price'] = x_test['Minimum_sales_price']*x_test['real']
```
